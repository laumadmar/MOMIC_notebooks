# Manhattans and QQplots

GWAS results are often displayed in a Manhattan plot with -log10(p-value) plotted against the position in the genome.

A p-value indicates the significance of the difference in frequency of the allele tested between cases and controls i.e. the probability that the allele is likely to be associated with the trait. 

```r
suppressPackageStartupMessages(library(qqman))
```

```r
path<-"/mnt/data/GWAS/output/build37/task4_assoc/"
results_assoc <- read.table(paste0(path,"dataset.b37.imputed.assoc.dosage.clean.rs.200kb.annot"), head=TRUE)
dim(results_assoc)
head(results_assoc)
```

## Manhattans

```r
# jpeg("path/Manhattan.jpeg")
manhattan(results_assoc, chr = "CHR", bp = "BP", p = "P", snp = "SNP", col = c("coral", "aquamarine4"), chrlabs = NULL, suggestiveline = -log10(1e-04), genomewideline = -log10(5e-08), highlight = NULL, logp = TRUE, main = "Manhattan plot", ylim = c(0, 10), cex = 0.6, cex.axis = 0.9)
#dev.off()
```

## QQ plots

```r
qq(results_assoc$P, main = "Q-Q plot of GWAS p-values : log")
```

```r
qq(results_assoc$P, main = "Q-Q plot of GWAS p-values : assoc")
```
